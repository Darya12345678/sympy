```python
from sympy import *
a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")
 
f_x = a * (x**3) + b * (x**2) + c * (x**1) + d
```

```python
f_x
```

```python
print (solve(f_x))
```

```python
from sympy.plotting import plot
g1 = plot(f_x,(x,-10,3), line_color = 'red', title = 'График функции')
```

```python
derivative = diff(f_x)
derivative
```

```python
g2 = plot(f_x, derivative,(x,-10,3), title = "Графики функции и производной функции", legend = True, xlabel = 'X', ylabel = 'Y', show = False)
g2[0].line_color = 'blue'
g2[1].line_color = 'green'
g2.show()
```

```python
a = solve(derivative > 0, x) 
print(f'{a}- интервалы возрастания функции')
```

```python
b = solve(derivative < 0, x)
print(f'{b} - интервалы убывания функции')
```
